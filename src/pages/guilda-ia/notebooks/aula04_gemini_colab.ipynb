## 1. Instalar SDK e Configurar API Key

O Google Gemini oferece **1 milhão de tokens/dia gratuito** — mais que suficiente para experimentar.

**Antes de continuar:**
1. Acesse https://aistudio.google.com
2. Faça login com sua conta Google
3. Clique em **"Get API Key"** e crie uma chave
4. No Colab, clique no ícone 🔑 (Secrets) na barra lateral
5. Adicione uma chave chamada `GEMINI_API_KEY` com o valor da sua API key

```python
# Instalar SDK do Google Gemini
!pip install google-genai

from google import genai
from google.genai import types
from google.colab import userdata

# Configurar API key (vem do Secrets do Colab)
client = genai.Client(api_key=userdata.get('GEMINI_API_KEY'))

print("✅ SDK configurado!")
```

## 2. Chamada Simples

Vamos perguntar algo direto ao modelo.

```python
# Mensagem simples — uma linha, direto ao ponto
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="Qual é a capital de Minas Gerais?"
)
print(response.text)
```

## 3. Chat com Memória

O método `chat` mantém o histórico automaticamente — o modelo lembra do que você disse antes.

```python
# Criar chat — o histórico é gerenciado pelo SDK
chat = client.chats.create(model="gemini-2.0-flash")

response1 = chat.send_message("Olá! Meu nome é Lucas e eu ensino IA.")
print(f"🤖: {response1.text}")

response2 = chat.send_message("Qual é o meu nome?")
print(f"🤖: {response2.text}")
```

## 4. System Prompt

O **system prompt** define o comportamento do modelo — como uma instrução geral antes da conversa.

```python
# Com system prompt — o modelo segue a instrução em todas as respostas
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="O que é uma lista em Python?",
    config=types.GenerateContentConfig(
        system_instruction="Você é um tutor de Python. Responda de forma didática, com exemplos curtos."
    )
)
print(response.text)
```

## 5. Parâmetros de Geração

- **temperature**: 0.0 (determinístico) → 2.0 (muito criativo)
- **max_output_tokens**: limite de tamanho da resposta
- **top_p** e **top_k**: controlam diversidade

```python
# Comparando temperatura baixa vs alta
for temp in [0.0, 0.7, 1.5]:
    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents="Conte uma piada sobre programação.",
        config=types.GenerateContentConfig(
            temperature=temp,
            max_output_tokens=60
        )
    )
    print(f"🌡️ temperature={temp}:")
    print(f"   {response.text}")
    print()
```

## 6. Tratamento de Erros

Erros são comuns com APIs. Os mais frequentes:
- **PermissionDenied**: API key inválida ou sem permissão
- **ResourceExhausted**: Limite de tokens excedido
- **NotFound**: Modelo não encontrado

```python
from google.api_core.exceptions import NotFound, PermissionDenied, ResourceExhausted

try:
    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents="Olá!"
    )
    print(response.text)
except PermissionDenied:
    print("❌ API key inválida ou sem permissão. Verifique sua chave no Secrets.")
except ResourceExhausted:
    print("⚠️ Limite de tokens excedido. Aguarde um momento e tente novamente.")
except NotFound:
    print("❌ Modelo não encontrado. Verifique o nome do modelo.")
except Exception as e:
    print(f"Erro: {e}")
```
